```python
import pandas as pd
import numpy as num
df=pd.read_csv("IRIS.csv")
df
```

```python
df['species']=df['species'].replace(['Iris-setosa','Iris-virginica','Iris-versicolor'],[1,2,3])
print(df)
```

```python
Sum_stats = df.groupby('species')[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].agg(['min', 'max', 'median', 'std'])
print(Sum_stats)
```
