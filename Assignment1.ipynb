```python
# for  normalization we used two method min-max and z-score method 
# min-max method it converted into range of 0-1
df['Pregnant']=(df['Pregnant']-df['Pregnant'].min())/(df['Pregnant'].max()-df['Pregnant'].min())
print(df['Pregnant'])
```

```python
# z-score it is used for converting columes value in the range of -3 to +3
df['Diastolic_BP']=(df['Diastolic_BP']-df['Diastolic_BP'].mean())/df['Diastolic_BP'].std()
print(df['Diastolic_BP'])
```

```python
# now normalization by using replace function
data={'Name':['Kiran','Shrutik','Sanket','Viraj'],'Roll_No':[101,102,103,104],'Education':['D','UG','PG','G'],'Grade':['A','A+','B','B+']}
ds=pd.DataFrame(data)
ds
```

```python
# now normalization by using replace function
ds['Education']=ds['Education'].replace(['D','UG','PG','G'],[1,2,3,4])
print(ds)
```

```python
# now normalization by using get_dummies() axis=1(column) adding here created grade values columns and then delete the original grade column
dummy=pd.get_dummies(ds.Grade)
merge=pd.concat([ds,dummy],axis=1)
merge.drop(['Grade'],axis=1)
```
